```python
import anndata
import numpy as np
import pandas as pd
import scanpy.api as sc
```

# Integrate the fetal gut samples and create an Anndata object

```python
fetal_metadata = pd.read_csv('/mnt/PROJECTS/Fetal_gut_project/sample_metadata/fetal_sample_tracker_02.txt', sep = '\t', index_col = 0)
```

```python
fetal_metadata
```

```python
path = '/mnt/Unprocessed_all/'
filenames = fetal_metadata.index
adatas = [sc.read_10x_mtx(path + filename + '/cellranger/', cache = True, make_unique = True) for filename in filenames]
for i in range(len(adatas)):
    adatas[i].obs['sample'] = fetal_metadata.index[i]
    for col in fetal_metadata.columns:
        adatas[i].obs[col] = fetal_metadata[col][i]
adata = adatas[0].concatenate(adatas[1:], batch_categories = fetal_metadata.index)
```
